# Which market should the battery participate in?

## Background

 - Multiple revenue streams exist to make a battery installation profitable in the UK. These include wholesale energy price arbitrage, peak shaving, and avoidance of distribution/transmission network costs.
 
 - All of these can currently generate revenues but often not enough to offer an interesting return on investment for investors. An additional revenue stream can come from providing balancing services to National Grid. The current market offers 4 different balancing services: Firm Frequency Response (FFR), Dynamic Containment (DC), Dynamic Regulation (DR), and Dynamic Moderation (DM). These services helps National Grid (the system operator) to continuously balance electricity supply and demand on the grid, stabilizing electricity frequency around the target value of 50 Hz.
 
 - Each service has a different "response curve" that requires the battery to output a certain percentage of its power capacity according to the system frequency at any given time
 
 - Every day, Zenobe has to decide which service to provide based on the value generated by each service (calculated using the auction clearing price and the cost of providing the service).
